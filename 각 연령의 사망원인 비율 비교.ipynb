```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
%matplotlib inline
```

```python
CODbyAge_man = pd.read_csv(r"C:\Users\digus\Desktop\db금융공모전\남성 연령별 사망자수(2001년~2018년).csv", engine='python')
```

```python
CODbyAge_man
```

```python
CODbyAge_man = CODbyAge_man.set_index("사망원인")
```

```python
CODbyAge_man
```

```python
plt.style.use('ggplot')   #Initialize the figure
palette = plt.get_cmap('rainbow')      # create a color palette
fig = plt.figure(figsize=(20,16))
fig.suptitle("남자 연령별 사망원인 비율", fontsize=30)
 # unicode minus를 사용하지 않기 위한 설정 (minus 깨짐현상 방지)
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
# multiple line plot
num=0
for column in CODbyAge_man:
    num+=1 
    plt.subplot(3,3, num)
    
    plt.plot(CODbyAge_man[column], marker='', color= palette(num), linewidth=1.9, alpha=0.9, label=column)
    #plt.xlim(0, 19)
    plt.ylim(0, 50)
    
    if num in range(7) :
        plt.xticks(ticks=[], labels = None)
    if num not in [1,4,7] :
        plt.tick_params(labelleft='off')
    if num in [7,8,9] :
        plt.xticks(fontsize=12, rotation=90)
        plt.xlabel('사망원인', fontsize= 12)
        
    plt.title(column, loc='center', fontsize=30, fontweight=0, color=palette(num))

    # Axis title
    
    plt.ylabel('사망률 (%)', fontsize=12)
   
```
