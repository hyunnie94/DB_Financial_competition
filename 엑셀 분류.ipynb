```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit_transform(man_cl)
#scaler.fit_transform(w)
feature = scaler.fit_transform(man_cl)
#feature2 = scaler.fit_transform(w)
```

```python
def elbow(X):
    sse = []

    for i in range(1,20):
        km = KMeans(n_clusters=i,algorithm='auto', random_state=42)
        km.fit(X)
        sse.append(km.inertia_)

    plt.plot(range(1,20), sse, marker='o')
    plt.xlabel('K')
    plt.ylabel('SSE')
    plt.show()
```

```python
elbow(feature)
```

```python
kmean_cl = KMeans(n_clusters=k, algorithm='auto')
kmeans = kmean_cl.fit(feature)
km_predict = kmean_cl.predict(feature)
km_predict
```
